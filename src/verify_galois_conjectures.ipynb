```python
def check_value_in(df: pl.DataFrame, index: int, values: List):
    col_values = df[f"a_{index:05d}"].unique().to_list()
    assert set(col_values) <= set(values), f"Invalid values in column a_{index:05d}: {col_values} not in {values}"
```

# Sextic

1. For $p \equiv 1 \pmod{6}$, we have

$$
    a_{p^2}(K) \in \begin{cases} \{0, 1, 3, 6, 21\} & \mathrm{Gal}(K/\mathbb{Q}) \simeq C_6 \\ \{0, 3, 6, 21\} & \mathrm{Gal}(K / \mathbb{Q}) \simeq S_3 \end{cases}
$$

2. For $p \equiv 5 \pmod{6}$, we have

$$
    a_{p^2}(K) \in \begin{cases} \{0, 3, 6, 21\} & \mathrm{Gal}(K/\mathbb{Q}) \simeq C_6 \\ \{0, 1, 3, 6, 21\} & \mathrm{Gal}(K / \mathbb{Q}) \simeq S_3 \end{cases}
$$

3. For $p \ge 5$, we have

$$
    a_{p^3}(K) \in \begin{cases} \{0, 1, 2, 4, 10, 56\} & \mathrm{Gal}(K/\mathbb{Q}) \simeq C_6 \\ \{0, 2, 4, 10, 56\} & \mathrm{Gal}(K / \mathbb{Q}) \simeq S_3 \end{cases}
$$

```python
df_sextic_galois = pl.read_csv(f"{current_dir}/data_nf/nf_6_galois.csv")
df_sextic_6T1 = df_sextic_galois.filter(pl.col("galois_label") == "6T1")
df_sextic_6T2 = df_sextic_galois.filter(pl.col("galois_label") == "6T2")
```

```python
for n in [5 ** 2, 7 ** 2, 5 ** 3]:
    print(f"all extensions for n = {n}")
    zeta_count(df_sextic_galois, n)
    print("6T1")
    zeta_count(df_sextic_galois, n, "galois_label", "6T1")
    print("6T2")
    zeta_count(df_sextic_galois, n, "galois_label", "6T2")
    print("-" * 20)
```

```python
N = 10000

# Square
for p in primes(int(N ** (1/2))):
    if p % 6 == 1:
        check_value_in(df_sextic_6T1, p**2, [0, 1, 3, 6, 21])
        check_value_in(df_sextic_6T2, p**2, [0, 3, 6, 21])
    elif p % 6 == 5:
        check_value_in(df_sextic_6T1, p**2, [0, 3, 6, 21])
        check_value_in(df_sextic_6T2, p**2, [0, 1, 3, 6, 21])

# Cube
for p in primes(int(N ** (1/3))):
    if p <= 3:
        continue
    check_value_in(df_sextic_6T1, p**3, [0, 1, 2, 4, 10, 56])
    check_value_in(df_sextic_6T2, p**3, [0, 2, 4, 10, 56])
```

# Decic

1. For $p \equiv 1 \pmod{10}$, we have

$$
    a_{p^2}(K) \in \begin{cases} \{0, 1, 3, 5, 15, 55\} & \mathrm{Gal}(K/\mathbb{Q}) \simeq C_{10} \\ \{0, 3, 5, 15, 55\} & \mathrm{Gal}(K / \mathbb{Q}) \simeq D_5 \end{cases}
$$

2. For $p \equiv 5 \pmod{6}$, we have

$$
    a_{p^2}(K) \in \begin{cases} \{0, 3, 5, 15, 55\} & \mathrm{Gal}(K/\mathbb{Q}) \simeq C_{10} \\ \{0, 1, 3, 5, 15, 55\} & \mathrm{Gal}(K / \mathbb{Q}) \simeq D_5 \end{cases}
$$

3. For $p \equiv 1 \pmod{10}$, we have

$$
    a_{p^3}(K) \in \begin{cases} \{0, 1, 4, 35, 220\} & \mathrm{Gal}(K/\mathbb{Q}) \simeq C_{10} \\ \{0, 4, 35, 220\} & \mathrm{Gal}(K / \mathbb{Q}) \simeq D_5 \end{cases}
$$

4. For $p \ge 7$, we have $a_{p^3}(K) \in \{0, 35, 220\}$.

```python
df_decic_galois = pl.read_csv(f"{current_dir}/data_nf/nf_10_galois_sq_cb_qu_100000.csv")
df_decic_10T1 = df_decic_galois.filter(pl.col("galois_label") == "10T1")
df_decic_10T2 = df_decic_galois.filter(pl.col("galois_label") == "10T2")
```

```python
for n in [11 ** 2, 19 ** 2, 11 ** 3, 7 ** 5]:
    print(f"n = {n}")
    print("all extensions")
    zeta_count(df_decic_galois, n)
    print("10T1")
    zeta_count(df_decic_galois, n, "galois_label", "10T1")
    print("10T2")
    zeta_count(df_decic_galois, n, "galois_label", "10T2")
    print("-" * 20)
```

```python
N = 100000

# Square
for p in primes(int(N ** (1/2))):
    if p % 10 == 1:
        check_value_in(df_decic_10T1, p**2, [0, 1, 3, 5, 15, 55])
        check_value_in(df_decic_10T2, p**2, [0, 3, 5, 15, 55])
    elif p % 10 == 9:
        check_value_in(df_decic_10T1, p**2, [0, 5, 15, 55])
        check_value_in(df_decic_10T2, p**2, [0, 1, 5, 15, 55])

# Cube 
for p in primes(int(N ** (1/3))):
    if p == 5:
        continue
    if p % 10 == 1:
        check_value_in(df_decic_10T1, p**3, [0, 1, 4, 35, 220])
        check_value_in(df_decic_10T2, p**3, [0, 4, 35, 220])
    else: 
        check_value_in(df_decic_10T1, p**3, [0, 35, 220])
        check_value_in(df_decic_10T2, p**3, [0, 35, 220])

# Fifth power
for p in primes(int(N ** (1/5))):
    if p == 5:
        continue
    check_value_in(df_decic_10T1, p**5, [0, 1, 2, 126, 2002])
    check_value_in(df_decic_10T2, p**5, [0, 2, 126, 2002])
```
