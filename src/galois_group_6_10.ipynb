### Classifying Galois group of Sextic and Decic extensions

```python
from utils import *

current_dir = globals()['_dh'][0]
```

### Sextic

- `6T1`: $C_6$
- `6T2`: $S_3$

```python
schema = {f"c_{i}": pl.Float64 for i in range(6)}
df_sextic_galois = pl.read_csv(f"{current_dir}/data_nf/nf_6_galois.csv", schema_overrides=schema)
```

```python
df_stats(df_sextic_galois, "galois_label")
```

```python
run_experiments(df_sextic_galois, "sextic_galois", ["6T1", "6T2"], current_dir, [2, 3])
```

```python
# Distribution of polynomial coefficients
df_sextic_6T1 = df_sextic_galois.filter(pl.col("galois_label") == "6T1")
df_sextic_6T2 = df_sextic_galois.filter(pl.col("galois_label") == "6T2")
for i in range(6):
    print(f"c_{i}:")
    print(f"Mean: 6T1: {df_sextic_6T1[f'c_{i}'].mean()}, 6T2: {df_sextic_6T2[f'c_{i}'].mean()}")
    print(f"Mediean: 6T1: {df_sextic_6T1[f'c_{i}'].median()}, 6T2: {df_sextic_6T2[f'c_{i}'].median()}")
```

```python
del df_sextic_galois, df_sextic_6T1, df_sextic_6T2
```

### Decic

- `10T1`: $C_{10}$
- `10T2`: $D_5$

```python
schema = {f"c_{i}": pl.Float64 for i in range(10)}
df_decic_galois = pl.read_csv(f"{current_dir}/data_nf/nf_10_galois.csv", schema_overrides=schema)
```

```python
df_stats(df_decic_galois, "galois_label")
```

```python
run_experiments(df_decic_galois, "decic_galois", ["10T1", "10T2"], current_dir, [2, 5])
```
