### Classifying Galois group of Quartic and Nonic extensions

```python
from utils import *

current_dir = globals()['_dh'][0]
```

### Quartic

- `4T1`: $C_4$
- `4T2`: $C_2^2$

```python
df_quartic_galois = pl.read_csv(f"{current_dir}/data_nf/nf_4_galois.csv")
```

```python
df_stats(df_quartic_galois, "galois_label")
```

```python
run_experiments(df_quartic_galois, "quartic_galois", ["4T1", "4T2"], current_dir, [2])
```

```python
# Distribution of polynomial coefficients
df_quartic_4T1 = df_quartic_galois.filter(pl.col("galois_label") == "4T1")
df_quartic_4T2 = df_quartic_galois.filter(pl.col("galois_label") == "4T2")
for i in range(4):
    print(f"c_{i}:")
    print(f"Mean: 4T1: {df_quartic_4T1[f'c_{i}'].mean()}, 4T2: {df_quartic_4T2[f'c_{i}'].mean()}")
    print(f"Mediean: 4T1: {df_quartic_4T1[f'c_{i}'].median()}, 4T2: {df_quartic_4T2[f'c_{i}'].median()}")
```

```python
del df_quartic_galois, df_quartic_4T1, df_quartic_4T2
```
