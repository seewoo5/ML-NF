### Classifying Galois group of octic extensions

```python
from utils import *

current_dir = globals()['_dh'][0]
```

### Octic

- `8T1`: $C_8$
- `8T2`: $C_4 \times C_2$
- `8T3`: $C_2^3$
- `8T4`: $D_4$
- `8T5`: $Q_8$

```python
schema = {f"c_{i}": pl.Float64 for i in range(8)}
df_octic_galois = pl.read_csv(f"{current_dir}/data_nf/nf_8_galois.csv", schema_overrides=schema)
```

```python
df_stats(df_octic_galois, "galois_label")
```

### Abelian extensions ($C_8$, $C_4 \times C_2$, $C_2^3$)

```python
df_octic_abelian = df_octic_galois.filter(pl.col("galois_label").is_in(["8T1", "8T2", "8T3"]))
df_stats(df_octic_abelian, "galois_label")
```

```python
run_experiments(df_octic_abelian, "octic_ab_galois", ["8T1", "8T2", "8T3"], current_dir, [2])
```

```python
# Distribution of polynomial coefficients
df_octic_8T1 = df_octic_abelian.filter(pl.col("galois_label") == "8T1")
df_octic_8T2 = df_octic_abelian.filter(pl.col("galois_label") == "8T2")
df_octic_8T3 = df_octic_abelian.filter(pl.col("galois_label") == "8T3")
for i in range(8):
    print(f"c_{i}:")
    print(f"Mean: 8T1: {df_octic_8T1[f'c_{i}'].mean()}, 8T2: {df_octic_8T2[f'c_{i}'].mean()}, 8T3: {df_octic_8T3[f'c_{i}'].mean()}")
    print(f"Mediean: 8T1: {df_octic_8T1[f'c_{i}'].median()}, 8T2: {df_octic_8T2[f'c_{i}'].median()}, 8T3: {df_octic_8T3[f'c_{i}'].median()}")
```

### Nonabelian extensions ($D_4, Q_8$)

```python
df_octic_nonabelian = df_octic_galois.filter(pl.col("galois_label").is_in(["8T4", "8T5"]))
df_stats(df_octic_nonabelian, "galois_label")
```

```python
run_experiments(df_octic_nonabelian, "octic_nab_galois", ["8T4", "8T5"], current_dir, [2])
```

```python
# Distribution of polynomial coefficients
df_octic_8T4 = df_octic_nonabelian.filter(pl.col("galois_label") == "8T4")
df_octic_8T5 = df_octic_nonabelian.filter(pl.col("galois_label") == "8T5")
for i in range(8):
    print(f"c_{i}:")
    print(f"Mean: 8T4: {df_octic_8T4[f'c_{i}'].mean()}, 8T5: {df_octic_8T5[f'c_{i}'].mean()}")
    print(f"Mediean: 8T4: {df_octic_8T4[f'c_{i}'].median()}, 8T5: {df_octic_8T5[f'c_{i}'].median()}")
```

### Abelian vs Nonabelian

```python
replace_dict = {
    "8T1": "abelian",
    "8T2": "abelian",
    "8T3": "abelian",
    "8T4": "nonabelian",
    "8T5": "nonabelian",
}
new_label_col = df_octic_galois["galois_label"].replace(replace_dict)
df_octic_ab_nab = df_octic_galois.clone()
df_octic_ab_nab = df_octic_ab_nab.replace_column(2, new_label_col)
```

```python
df_stats(df_octic_ab_nab, "galois_label")
```

```python
run_experiments(df_octic_ab_nab, "octic_ab_nab_galois", ["abelian", "nonabelian"], current_dir, [2])
```

```python
# Distribution of polynomial coefficients
for i in range(8):
    print(f"c_{i}:")
    print(f"Mean: ab: {df_octic_abelian[f'c_{i}'].mean()}, nab: {df_octic_nonabelian[f'c_{i}'].mean()}")
    print(f"Mediean: ab: {df_octic_abelian[f'c_{i}'].median()}, nab: {df_octic_nonabelian[f'c_{i}'].median()}")
```

```python
del df_octic_galois, df_octic_8T1, df_octic_8T2, df_octic_8T3, df_octic_8T4, df_octic_8T5
del df_octic_ab_nab, df_octic_abelian, df_octic_nonabelian
```
